```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib

%matplotlib inline

plt.style.use('seaborn-whitegrid')

tfont = {'fontname':'Serif'}
```

```python
data = pd.read_csv('C:/Users/samuel/Google Drive/Elise Projects/Friend or Foe/FoF_data/ae_plot_data.csv', names = ['ID', 'Sound', 'No Sound', 'SE'],
                  skiprows = 1)

# data = data[data['Experiment'] == 'first']
# data['No Sound'] = data['No Sound'].astype(float)
# data['Sound'] = data['Sound'].astype(float)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10.5))
fig.patch.set_facecolor('#FFFFFF')
fig.tight_layout(pad=6)

axs = [ax1, ax2]
for i, ax in enumerate(axs):
    ax.bar(range(1, 48), data.iloc[:, i + 1], width = 0.5)
    ax.errorbar(range(1, 48), data.iloc[:, i + 1], data['SE'], color = 'black', ls = 'none', capsize = 3, lw = 1)
    ax.tick_params(direction='inout', length=10, width=1, color='lightgray', axis='x', 
                   grid_color='gray', labelsize=12)
    ax.tick_params(direction='in', length=0, width=0, axis='y', labelsize=12)
    ax.set_yticks([-0.5, 0, 0.5, 1, 1.5])
    ax.set_xticks([])
    ax.set_xlabel('Participant #', fontsize=20, **tfont)
    
    ax.set_ylabel('After-effect size', fontsize=20, **tfont)
    ax.set_title(data.columns[i + 1], fontsize=24, **tfont)
    ax.grid(axis='x')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)

    for tick in ax.get_yticklabels():
       tick.set_fontname('Serif')
    for tick in ax.get_xticklabels():
       tick.set_fontname('Serif')

    
plt.gcf().text(0.03, 0.95, 'A', fontsize=24, **tfont)
plt.gcf().text(0.03, 0.48, 'B', fontsize=24, **tfont)

matplotlib.pyplot.subplots_adjust(hspace=0.45)
```
